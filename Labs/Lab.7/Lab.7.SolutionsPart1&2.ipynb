```python
# part a

def pairplot_a(matrix1, matrix2, vars=None, figsize=(10, 10), marker='o', color1='b', color2='r', alpha=0.5):
    if vars is None:
        vars = matrix1.select_dtypes(include=['number']).columns.tolist()

    num_vars = len(vars)

    fig,axes = plt.subplots(num_vars, num_vars, figsize=figsize)

    for i, var1 in enumerate(vars):
        for j, var2 in enumerate(vars):
            if i == j:
                axes[i,j].hist(matrix1[var1], bins=20, color=color1, alpha=alpha)
                axes[i,j].hist(matrix2[var1], bins=20, color=color2, alpha=alpha)
                axes[i,j].set_xlabel(var1)
                axes[i,j].set_ylabel('Frequency')
            else:
                axes[i,j].scatter(matrix1[var2], matrix1[var1], marker=marker, color=color1, alpha=alpha)
                axes[i,j].scatter(matrix2[var2], matrix2[var1], marker=marker, color=color2, alpha=alpha)
                axes[i,j].set_xlabel(var2)
                axes[i,j].set_ylabel(var1)

    plt.tight_layout()
    plt.show()
    
pairplot_a(df_sig, df_bkg, vars=["l_1_pT", "l_1_eta","l_1_phi"])
```

```python
# part b
def pairplot_b(matrix1, matrix2, vars=None, figsize=(10, 10), marker='o', color1='b', color2='r', alpha=0.5):
  if vars is None:
    vars = matrix1.select_dtypes(include=['number']).columns.tolist()

  num_vars = len(vars)

  fig, axes = plt.subplots(num_vars, num_vars, figsize=figsize)

  # Pre-compute histograms for faster plotting
  hists = {}
  for var in vars:
    counts1, bins1 = np.histogram(matrix1[var], bins=20)
    counts2, bins2 = np.histogram(matrix2[var], bins=20)
    hists[var] = (counts1, bins1)
    hists[var] = (counts2, bins2)

  for i, var1 in enumerate(vars):
    for j, var2 in enumerate(vars):
      if i == j:
        # Use ax.hist with unpacked data (counts & bins)
        axes[i, j].hist(hists[var1][0], bins=hists[var1][1])
        axes[i, j].set_xlabel(var1)
        axes[i, j].set_ylabel('Frequency')
      else:
        axes[i, j].scatter(matrix1[var2], matrix1[var1], marker=marker, color=color1, alpha=alpha)
        axes[i, j].scatter(matrix2[var2], matrix2[var1], marker=marker, color=color2, alpha=alpha)
        axes[i, j].set_xlabel(var2)
        axes[i, j].set_ylabel(var1)

  plt.tight_layout()
  plt.show()
    
pairplot_b(df_sig, df_bkg, vars=["l_1_pT","l_1_eta","l_1_phi","cos_theta_r1"])
```

```python
# 5.3
def compare_significance(df_sig, df_bkg,
                         obs_name,
                         scenarios, bins=100, log=False):
    
    TPR, bins_sig = compute_rate(df_sig[obs_name], bins=bins)
    FPR, _ = compute_rate(df_bkg[obs_name], bins=bins_sig)
    
    max_sigs = {}
    table = []
    
    plt.figure(figsize=(12, 8))
    for name, (n_sig_expected, n_bkg_expected) in scenarios.items():
        n_sig_expected_prime = n_sig_expected * TPR
        n_bkg_expected_prime = n_bkg_expected * FPR
        
        significance = n_sig_expected_prime / np.sqrt(np.maximum(n_sig_expected_prime + n_bkg_expected_prime, 1e-10))
        
        plt.step(bins_sig[:-1], significance, label=f"{name} ({n_sig_expected}, {n_bkg_expected})")
        
        max_i = np.argmax(significance)
        max_sigs[name] = (max_i, n_sig_expected_prime[max_i], n_bkg_expected_prime[max_i], significance[max_i], bins_sig[max_i])
        table.append((name, n_sig_expected, n_bkg_expected,
                      TPR[max_i], FPR[max_i],
                      n_sig_expected_prime[max_i], n_bkg_expected_prime[max_i], significance[max_i], bins_sig[max_i], max_i))
    
    if log:
        plt.yscale("log")
    plt.xlabel("Observable Threshold (x_c)")
    plt.ylabel("Significance (Ïƒ_S')")
    plt.title(f"Significance vs Threshold for {obs_name}")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()
    
    # table
    display(HTML(tabulate.tabulate(table, tablefmt='html',
                                   headers=["Name", "N_S", "N_B", "TPR", "FPR", "N_S'", "N_B'", "Ïƒ_S'", "x_c", "bin i"])))
    return max_sigs


scenarios = {
    "Scenario 1": (10, 100),
    "Scenario 2": (100, 1000),
    "Scenario 3": (1000, 10000),
    "Scenario 4": (10000, 100000)
}

# compare
results = compare_significance(df_sig, df_bkg, "M_TR_2", scenarios)
```

```python
# 5.3
def compute_rate(data, bins):
    hist, bin_edges = np.histogram(data, bins=bins, density=False)
    cumulative_counts = np.cumsum(hist[::-1])[::-1]  # UNO Reverse
    total_counts = cumulative_counts[0]
    rates = cumulative_counts / total_counts if total_counts > 0 else np.zeros_like(cumulative_counts)
    return rates, bin_edges

def compare_significance(df_sig, df_bkg, obs_name, scenarios, bins=100, log=False):
   
    TPR, bins_sig = compute_rate(df_sig[obs_name], bins=bins)
    FPR, _ = compute_rate(df_bkg[obs_name], bins=bins_sig)
    
    max_sigs = {}
    table = []
    
    plt.figure(figsize=(12, 8))
    for name, (n_sig_expected, n_bkg_expected) in scenarios.items():
        n_sig_expected_prime = n_sig_expected * TPR
        n_bkg_expected_prime = n_bkg_expected * FPR
        
        # Compute significance
        significance = n_sig_expected_prime / np.sqrt(np.maximum(n_sig_expected_prime + n_bkg_expected_prime, 1e-10))
        
        plt.step(bins_sig[:-1], significance, label=f"{name} ({n_sig_expected}, {n_bkg_expected})")
        
        max_i = np.argmax(significance)
        max_sigs[name] = (max_i, n_sig_expected_prime[max_i], n_bkg_expected_prime[max_i], significance[max_i], bins_sig[max_i])
        table.append((name, n_sig_expected, n_bkg_expected,
                      TPR[max_i], FPR[max_i],
                      n_sig_expected_prime[max_i], n_bkg_expected_prime[max_i], significance[max_i], bins_sig[max_i], max_i))
    
    if log:
        plt.yscale("log")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()
    
    display(HTML(tabulate.tabulate(table, tablefmt='html',
                                   headers=["Name", "N_S", "N_B", "TPR", "FPR", "N_S'", "N_B'", "Ïƒ_S'", "x_c", "bin i"])))
    return max_sigs

scenarios = {
    "Scenario 1": (10, 100),
    "Scenario 2": (100, 1000),
    "Scenario 3": (1000, 10000),
    "Scenario 4": (10000, 100000)
}

results = compare_significance(df_sig, df_bkg, "M_TR_2", scenarios)
```

```python
# 7.1
def compute_rate(d,bins=100):
    hist,bins_=np.histogram(d,bins=bins,density=True)
    R = np.cumsum(hist[::-1])[::-1] * (bins_[1]-bins_[0])
    return R,bins_

def AUC(TPR,FPR):
    n = FPR.shape[0]
    return np.sum((FPR[0:n-1]-FPR[1:]) * (TPR[0:n-1]+TPR[1:])/2)
```

```python
# 8.1

df_sig_0 = df_sig.drop("signal",axis=1)
df_bkg_0 = df_bkg.drop("signal",axis=1)

# mean of each feature
m_s= np.mean(df_sig_0,axis=0)
m_b= np.mean(df_bkg_0,axis=0)

# (m2-m1)
delta = m_s-m_b
delta=np.matrix(m_s-m_b).transpose() #turn delta into matrix to calculate

# Sb
S_B= delta*delta.transpose()


# delta
delta_s=np.matrix(df_sig_0-m_s).transpose()
delta_b=np.matrix(df_bkg_0-m_b).transpose()

# signal
S_W_s= delta_s*delta_s.transpose()

# background
S_W_b= delta_b*delta_b.transpose()

# Sw
S_W=S_W_s+S_W_b


# calculate w
S_W_inv = np.linalg.inv(S_W)
w = S_W_inv * np.matrix(m_b - m_s).transpose()

# Projections
output_s = np.array(df_sig_0 @ w).flatten()
output_b = np.array(df_bkg_0 @ w).flatten()

# Histogram
bins = np.histogram_bin_edges(np.concatenate([output_s, output_b]), bins="auto")
plt.hist(output_s, bins=bins, label="Signal", alpha=0.5)
plt.hist(output_b, bins=bins, label="Background", alpha=0.5)
plt.legend()
plt.xlabel("Projection onto w")
plt.ylabel("Frequency")
plt.title("Signal vs Background Projections")
plt.show()
```
