1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value):
        self.current_value=0
        self.max_value = max_value
    def increment(self):
        if self.current_value < self.max_value:
            self.current_value += 1
        else:
            print(f"Cannot increment beyond the maximum value of {self.max_value}.")
            
    def reset(self):
        self.current_value=0
        
    def display(self):
        print(f"Counter Value: {self.current_value}")
    
```

```python
###check

counter = Counter(5)

counter.increment()
counter.display()

#testing beyond the range
for _ in range(7):
    counter.increment()
    counter.display()
    
counter.reset()
counter.display()
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
class Shape:
    def area(self):
        return "Area method not yet implemented; override"
    
    def perimeter(self):
        return "Perimeter method not yet implemented; override"
    
    def display(self):
        return "Display method not yet implemented; override"
    
    def perimeter_points(self):
        return "Perimeter points method not yet implemented; override"
    
    def inside_scoop(self, x, y):
        return "Inside coordinates have not yet been implemented; override"

    def overlaps(self, other):
        return "Overlap method not yet implemented; override"


class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * (self.__length + self.__width)
  
    def display(self):
        print(f"Rectangle: length={self.__length}, width={self.__width}, corner=({self.__x}, {self.__y})")
     
    def perimeter_points(self):
        points = []
        for i in range(16):
            if i < 4:
                x = self.__x + (i / 4) * self.__length
                y = self.__y
            elif i < 8:
                x = self.__x + self.__length
                y = self.__y + ((i - 4) / 4) * self.__width
            elif i < 12:
                x = self.__x + (12 - i) / 4 * self.__length
                y = self.__y + self.__width
            else:
                x = self.__x
                y = self.__y + (16 - i) / 4 * self.__width
            points.append((x, y))
        return points
    
    def inside_scoop(self, x, y):
        return self.__x <= x <= (self.__x + self.__length) and self.__y <= y <= (self.__y + self.__width)

    def overlaps(self, other):
        if isinstance(other, Rectangle):
            return not (self.__x + self.__length < other.__x or
                        other.__x + other.__length < self.__x or
                        self.__y + self.__width < other.__y or
                        other.__y + other.__width < self.__y)
        elif isinstance(other, Circle):
            distance_x = abs(other._Circle__x - (self.__x + self.__length / 2))
            distance_y = abs(other._Circle__y - (self.__y + self.__width / 2))
            if distance_x > (self.__length / 2 + other._Circle__radius):
                return False
            if distance_y > (self.__width / 2 + other._Circle__radius):
                return False
            return True
        elif isinstance(other, Triangle):
            return (self.inside_scoop(other._Triangle__x, other._Triangle__y) or
                    self.inside_scoop(other._Triangle__x + other._Triangle__a, other._Triangle__y) or
                    self.inside_scoop(other._Triangle__x + other._Triangle__a / 2, other._Triangle__y + other._Triangle__b))
        return False


class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def area(self):
        return 3.14 * (self.__radius ** 2)

    def perimeter(self):
        return 2 * 3.14 * self.__radius

    def display(self):
        print(f"Circle: radius={self.__radius}, center=({self.__x}, {self.__y})")
        
    def perimeter_points(self):
        points = []
        for i in range(16):
            angle = (i / 16) * (2 * 3.14)  # Angle in radians
            x = self.__x + self.__radius * (3.14 ** 0.5) * (angle)  # Adjust with cos and sin
            y = self.__y + self.__radius * (3.14 ** 0.5) * (angle)
            points.append((x, y))
        return points
    
    def inside_scoop(self, x, y):
        return ((x - self.__x) ** 2 + (y - self.__y) ** 2) <= (self.__radius ** 2)

    def overlaps(self, other):
        if isinstance(other, Circle):
            distance = ((self.__x - other.__x) ** 2 + (self.__y - other.__y) ** 2) ** 0.5
            return distance < (self.__radius + other.__radius)
        elif isinstance(other, Rectangle):
            return other.overlaps(self) 
        elif isinstance(other, Triangle):
            return other.overlaps(self)
        return False


class Triangle(Shape):
    def __init__(self, a, b, c, x, y):
        self.__a = a
        self.__b = b
        self.__c = c
        self.__x = x
        self.__y = y

    def area(self): 
        s = (self.__a + self.__b + self.__c) / 2
        area = (s * (s - self.__a) * (s - self.__b) * (s - self.__c)) ** 0.5
        return area
    
    def perimeter(self):
        return self.__a + self.__b + self.__c
    
    def display(self):
        print(f"Triangle: sides=({self.__a}, {self.__b}, {self.__c}), vertex=({self.__x}, {self.__y})")
        
    def perimeter_points(self):
        points = []
        for i in range(16):
            if i < 5:
                x = self.__x + (i / 4) * self.__a
                y = self.__y
            elif i < 10:
                x = (self.__x + self.__a) - (i - 4) / 5 * (self.__a / 2)
                y = self.__y + ((i - 4) / 5) * self.__b
            else:
                x = self.__x + ((10 - i) / 5) * (self.__a / 2)
                y = self.__y + (5 / 5) * self.__b
            points.append((x, y))
        return points 
    
    def inside_scoop(self, x, y):
        def area(x1, y1, x2, y2, x3, y3):
            return abs((x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)) / 2.0)
        A = area(self.__x, self.__y, self.__x + self.__a, self.__y, self.__x + self.__a / 2, self.__y + self.__b)
        A1 = area(x, y, self.__x, self.__y, self.__x + self.__a, self.__y)
        A2 = area(x, y, self.__x + self.__a, self.__y, self.__x + self.__a / 2, self.__y + self.__b)
        A3 = area(x, y, self.__x + self.__a / 2, self.__y + self.__b, self.__x, self.__y)
        return A == A1 + A2 + A3

    def overlaps(self, other):
        if isinstance(other, Triangle):
            return (self.inside_scoop(other.__x, other.__y) or
                    self.inside_scoop(other.__x + other.__a, other.__y) or
                    self.inside_scoop(other.__x + other.__a / 2, other.__y + other.__b))
       
```

```python
class RasterDrawing:
    def __init__(self, canvas):
        self.canvas = canvas
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def remove_shape(self, shape):
        self.shapes.remove(shape)

    def paint(self):
        self.canvas.clear_canvas() 
        for shape in self.shapes:
            self.draw_shape(shape)

    def draw_shape(self, shape):
        if isinstance(shape, Rectangle):
            self.draw_rectangle(shape)
        elif isinstance(shape, Circle):
            self.draw_circle(shape)
        elif isinstance(shape, Triangle):
            self.draw_triangle(shape)

    def draw_rectangle(self, rectangle):
        x, y = rectangle._Rectangle__x, rectangle._Rectangle__y
        width, height = rectangle._Rectangle__length, rectangle._Rectangle__width
        for i in range(width):
            for j in range(height):
                self.canvas.set_pixel(x + i, y + j, '*')

    def draw_circle(self, circle):
        x_center, y_center = circle._Circle__x, circle._Circle__y
        radius = circle._Circle__radius
        for x in range(x_center - radius, x_center + radius + 1):
            for y in range(y_center - radius, y_center + radius + 1):
                if (x - x_center) ** 2 + (y - y_center) ** 2 <= radius ** 2:
                    self.canvas.set_pixel(x, y, '*')

    def draw_triangle(self, triangle):
        x, y = triangle._Triangle__x, triangle._Triangle__y
        a = triangle._Triangle__a
        height = triangle._Triangle__b 

        points = [(x, y), (x + a, y), (x + a / 2, y + height)]

        for i in range(3):
            start = points[i]
            end = points[(i + 1) % 3]
            self.canvas.line(start[0], start[1], end[0], end[1], char='*')


```

```python
### check

from paint import Canvas, Rectangle, Circle

canvas = Canvas(20, 20)

drawing = RasterDrawing(canvas)

rectangle = Rectangle(5, 3, 2, 2)
circle = Circle(4, 10, 10)

drawing.add_shape(rectangle)
drawing.add_shape(circle)

print("Initial Drawing:")
drawing.paint()
canvas.display()

drawing.remove_shape(circle)

print("Modified Drawing (after removing circle):")
drawing.paint()
canvas.display()
```

   * Overload `__repr__` functions of all objects to return strings of the python code that would construct the object.
   
   * In the save method of raster drawing class, store the representations into the file.
   * Write a loader function that reads the file and uses `eval` to instantiate the object.

For example:

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    with open(filename, "r") as f:
        tmp = eval(f.read())
    return tmp
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for _ in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x, x + w):
            self.set_pixel(i, y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y, y + h):
            self.set_pixel(x, i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        # Handle vertical lines
        if x1 == x2:
            for y in range(min(y1, y2), max(y1, y2) + 1):
                self.set_pixel(y, x1, **kargs)
            return

        # Handle horizontal lines
        if y1 == y2:
            for x in range(min(x1, x2), max(x1, x2) + 1):
                self.set_pixel(y1, x, **kargs)
            return
        
        # Handle regular lines
        slope = (y2 - y1) / (x2 - x1)
        for x in range(int(x1), int(x2) + 1):
            y = int(slope * (x - x1) + y1)
            self.set_pixel(y, x, **kargs)

    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```
